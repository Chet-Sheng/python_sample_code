```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
def linear_2d_function(x_1, x_2):
    y=1*x_1-2*x_2
    return y
```

```python
x1=np.array(range(-10,10))
x2=np.array(range(-10,10))
y=linear_2d_function(x1,x2)
```

```python
ax.scatter(x1,x2,y, cmap=plt.cm.Spectral)
plt.show()
```
